## Numpy Continuation

### Array Operations

```python
import numpy as np
```

```python
arr_1 = np.array([[5,6,8], [9,12,16], [-7,-10,-13]])
```

```python
arr_2 = np.ones(3, dtype=int)
```

#### Addition

```python
arr_1 + arr_2
```

#### Subtraction

```python
arr_1 - arr_2
```

#### Multiplication

```python
arr_3 = np.array([0,10,20])
```

```python
arr_1 * arr_3
```

#### Division

```python
arr_4 = np.array([10,20,30])
```

```python
arr_1 / arr_4
```

`sum()` is used to find the sum of the elements in the array.

```python
arr_4.sum()
```

To add rows or columns in a nD array, you would specify the axis.

You can sum the elemets over rows with

```python
arr_1.sum(axis=0)
```

Sum elements over columns with

```python
arr_1.sum(axis=1)
```

#### Broadcasting

Broadcasting is simply an operation between an _array_ and a _single number_ or operations between arrays of different sizes. Also, an operation between a _vector_ and a _scalar_.
Broadcasting is a mechanism that allows NumPy to perform operations on arrays of different shapes

```python
arr_1
```

```python
arr_1 * -10
```

```python
arr_5 = np.array([2,4,6])
```

```python
arr_1 * arr_5
```

The dimensions of your array must be compatible, for example, when the dimensions of both arrays are equal or when one of them is 1. If the dimensions are not compatible, you will get a `ValueError`.

#### Mean
`mean` is used to find the mean of the elements in an array.

```python
avg_arr = np.array([[10,24,48],
                    [18,48,96],
                    [-14,-40,-78]])
```

```python
avg_arr.mean()
```

You can specify the _axis_ on which you want to find the mean.

```python
avg_arr.mean(axis=0)

# Mean over the rows
```

```python
avg_arr.mean(axis=1)

# Mean over the columns
```

#### Product

`prod` is used to get the result of multiplying the elements together

```python
prod_arr = np.array([[10,24,48],
                    [18,48,96],
                    [-14,-40,-78]])
```

```python
prod_arr.prod()
```

```python
prod_arr.prod(axis=0) # Row Products
```

```python
prod_arr.prod(axis=1) # Column Products
```
