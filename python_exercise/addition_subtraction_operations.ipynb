# Addition and Subtraction Operations

## Import

```python
import numpy as np
```

## Create Matrices

```python
m = np.array([[2,3,4],[6,7,8]])
```

```python
n = np.array([[64,55,47],[11,34,23]])
```

## Addition of matrices

```python
m + n
```

## Subtract Matrices

```python
n - m
```

```python
m - n
```

## Vectors

```python
v1 = np.array([3,7,11])
```

```python
v2 = np.array([19,15,-11])
```

## Vector Addition

```python
v1 + v2
```

## Vector Difference

```python
v1 - v2
```
