# Transposing

```python
import numpy as np
```

## Matrices

```python
m = np.array([[2,3,4],[6,7,8]])
m
```

```python
m.T
```

## Vectors

```python
v = np.array([2,3,4,5])
v
```

```python
v_reshaped = v.reshape(4,1)
v_reshaped
```

```python
v_reshaped.T
```

## Scalars

```python
s = np.array([9])
s
```

```python
s.T
```
