# Creating Tensors

# Imports

```python
import numpy as np
```

# Create matrices

```python
m = np.array([[2,3,4],[6,7,8]])
m
```

```python
n = np.array([[64,55,47],[11,34,23]])
n
```

# Create a tensor from the matrices

```python
t = np.array([m,n])
```

```python
t
```
