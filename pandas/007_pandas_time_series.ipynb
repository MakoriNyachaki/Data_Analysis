# Time Series

 A simple, powerful, and efficient functionality for performing resampling operations during frequency conversion (e.g., converting secondly data into 5-minutely data) in pandas.
 1. `Series.tz_localize()` localizes a time series to a time zone.
 2. `Series.tz_convert()` converts a timezones aware time series to another time zone.

```python
import numpy as np
import pandas as pd
```

```python
date_series = pd.date_range("1/1/2020", periods=1000, freq="s")
ts = pd.Series(np.random.randint(0, 5000, len(date_series)), index=date_series)
```

```python
ts.resample("5min").sum()
```

```python
# Localizing time

localize_time = pd.date_range("3/6/2012 00:00", periods=5, freq="D")
ts1 = pd.Series(np.random.randn(len(localize_time)), localize_time)
ts1
```

```python
ts_utc = ts1.tz_localize("UTC")
```

```python
ts_utc
```

```python
# Convert timezones
ts_utc.tz_convert("US/Eastern")
```

```python
ts_utc
```

```python
localize_time + pd.offsets.BusinessDay(5)
```
