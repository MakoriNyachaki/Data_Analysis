# Pandas

### Importing Pandas

```python
import numpy as np
import pandas as pd
```

### Data Structures in Pandas

`Series`:  one-dimensional labeled array holding data of any type.

`DataFrame`: a two-dimensional data structure that holds data like a two-dimension array or a table with rows and columns.

### Creating Objects
Creating a `Series` by passing a list of values, letting pandas create a default `RangeIndex`.

```python
pd_series = pd.Series([1,3,5,7,np.nan,11,13])
```

```python
pd_series
```

Creating a `DataFrame` by passing a NumPy array with a datetime index using `date_range()` and labeled columns:

```python
date_pd = pd.date_range('20160101', periods=10)
```

```python
date_pd
```

```python
arr_df =  pd.DataFrame(np.random.randn(10, 4), index=date_pd, columns=list("ABCD"))
```

```python
arr_df
```

Creating a `DataFrame` from a dictionary where the keys are the column labels and the values are column values.

```python
dict_df = pd.DataFrame(
    {
        "A": 1.0,
        "B": pd.Timestamp("20130102"),
        "C": pd.Series(1, index=list(range(4)), dtype="float32"),
        "D": np.array([3] * 4, dtype="int32"),
        "E": pd.Categorical(["test", "train", "test", "train"]),
        "F": "foo",
    }
)
```

```python
dict_df
```

```python
dict_df.dtypes
```

### Viewing Data
`DataFrame.head()`, `DataFrame.tail()`

`DataFrame.head()`: is used to view the top rows of the frame

`DataFrame.tail()`: is used to view the bottom rows of the frame.

```python
arr_df.head()
```

```python
arr_df.tail()
```

`DataFrame.index `: to show the indices of the DataFrame

`DataFrame.columns`: to display the columns of the DataFrame

```python
arr_df.index
```

```python
arr_df.columns
```

`DataFrame.to_numpy()`: Used to return a Numpy represenation of the underlying data with or without index or column labels.

```python
arr_df.to_numpy()
```

_**Note**_

NumPy arrays have one `dtype` for the entire array while pandas `DataFrames` have one dtype per column. When you call `DataFrame.to_numpy()`, pandas will find the NumPy dtype that can hold all of the dtypes in the DataFrame. If the common data type is `object`, `DataFrame.to_numpy()` will require copying data.

```python
arr_df.dtypes
```

```python
arr_df.to_numpy()
```

### Getitem([])

```python
arr_df['A']
```

```python
# If the label only contains letters, numbers, and underscores,
# you can alternatively use the column name attribute
arr_df.A
```

```python
arr_df['20160103':'20160106']
```

##### Selection by Label
Selecting a row matching a label

```python
arr_df.loc[date_pd[0]]
```

Selecting all rows (:) with a select column labels:

```python
arr_df.loc[:, ['A', 'B']]
```

For label slicing, both endpoints are included

```python
arr_df.loc['20160102':'20160107', ['A', 'C']]
```

Selecting a single row and column label returns a scalar

```python
arr_df.loc[date_pd[0], 'C']
```

```python
# For getting fast access to a scalar (equivalent to the prior method)
arr_df.at[date_pd[0], 'C']
```
