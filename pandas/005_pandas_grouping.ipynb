# Grouping
We use `groupby` to group pandas objects. Grouping refers to one of the following:
1. *Splitting:* Placing data in groups based on a criterion.
2. *Applying:* Apply different functions to each group independently.
3. *Combining:* Combine the result into a data structure.

```python
import numpy as np
import pandas as pd
```

```python
group_df = pd.DataFrame({
    "A":["foo", "bar", "bar", "foo", "foo", "bar", "bar", "foo"],
    "B":["one", "two", "two", "three", "four", "one", "three", "one"],
    "C": np.random.randn(8),
    "D": np.random.randn(8)
})
```

```python
group_df
```

Grouping by a column label, selecting column labels, and then applying the `DataFrameGroupBy.sum()` function to the resulting groups:

```python
# Grouping by a single column
group_df.groupby("A")[["C","D",]].sum()
```

```python
# Grouping by multiple columns
group_df.groupby(["A", "B"])[["C", "D"]].sum()
```
