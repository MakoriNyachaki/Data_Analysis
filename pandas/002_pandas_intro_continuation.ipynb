### DataFrames

```python
data = {'Name':['Amy White', 'Jack Stewart', 'Richard Lauderdale', 'John Owen'],
        'Age':[50,43,54,65],
        'Working Experience (Yrs.)': [5,8,3,2]
       }
df = pd.DataFrame(data)
print(df)
```

```python
data_1 = [{'Name':'Amy White', 'Age':50, 'Working Experience (Yrs.)':5},
         {'Name':'Jack Stewart', 'Age':43, 'Working Experience (Yrs.)':8},
         {'Name':'Richard Lauderdale', 'Age':54, 'Working Experience (Yrs.)':3},
         {'Name':'John Owen', 'Age':65, 'Working Experience (Yrs.)':2}
         ]
df_1 = pd.DataFrame(data_1)
print(df_1)
```

```python
# Creating a DataFrame from Pandas Series

names = pd.Series(['Amy White', 'Jack Stewart', 'Richard Lauderdale', 'John Owen'])
age = pd.Series([50, 43, 54, 65])
work_experience_yrs = pd.Series([5, 8, 3, 2])

data = {"Names": names, "Age": age, "Working Experience (Yrs.)": work_experience_yrs}
df_2 = pd.DataFrame(data)
print(df_2)
```

```python
# Creating a DataFrame from a list of lists.

data = [['Amy White', 50, 5], ['Jack Stewart', 43, 8], ['Richard Lauderdale', 54, 3], ['John Owen', 65, 2]]
df_3 = pd.DataFrame(data, columns = ['Name', 'Age', 'Working Experience (Yrs.)'])
print(df_3)
```

```python
# Adding label-based indices

data = [['Amy White', 50, 5], ['Jack Stewart', 43, 8], ['Richard Lauderdale', 54, 3], ['John Owen', 65, 2]]
df_3 = pd.DataFrame(data, columns = ['Name', 'Age', 'Working Experience (Yrs.)'], index = [1, 2, 3, 4])
print(df_3)
```
