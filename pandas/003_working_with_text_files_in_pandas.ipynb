# Files

A file is any item that can be created, stored, or deleted by the user or operating system.

### CSV

`DataFrame.to_csv()` is used to write into a csv file.

`read_csv()` is used to read a csv file.

```python
import numpy as np
import pandas as pd
```

```python
csf_df = pd.DataFrame(np.random.randint(0, 100, (1000, 5)))
```

```python
# Saving to a csv file
csf_df.to_csv("csv-dataframe.csv")
```

```python
# Reading a CSV file
pd.read_csv("csv-dataframe.csv")
```

### Parquet
`DataFrame.to_parquet()` is used to write a parquet file.

`pandas.read_parquet()` is used to read a parquet file.

```python
par_df = pd.DataFrame(np.random.randint(0, 100, (1000, 10)))
```

```python
# Writing into a parquet file
par_df.to_parquet("par-data.parquet")
```

```python
# Reading a parquet file
pd.read_parquet("par-data.parquet")
```

### Excel
`DataFrame.to_excel()` is used to write into an excel file.

`read_excel()` is used to read from an Excel file.

```python
excel_df = pd.DataFrame(np.random.randint(0, 100, (10000, 20)))
```

```python
excel_df.to_excel("excel-raw-data.xlsx", sheet_name="Saved-From-Pandas")
```

```python
pd.read_excel("excel-raw-data.xlsx", "Saved-From-Pandas", index_col=None, na_values=["NA"])
```
