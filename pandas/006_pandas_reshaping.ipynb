# Reshaping
### Stack
`stack()` method compresses a level in the `DataFrame`'s  columns.

```python
import numpy as np
import pandas as pd
```

```python
arrays = [
   ["bar", "bar", "baz", "baz", "foo", "foo", "qux", "qux"],
   ["one", "two", "one", "two", "one", "two", "one", "two"],
]
```

```python
index = pd.MultiIndex.from_arrays(arrays, names=['first','second'])
stack_df = pd.DataFrame(np.random.randn(8,2), index=index, columns=["A","B"])
```

```python
stack_df
```

```python
stack_df_2 = stack_df[:4]
```

```python
stack_df_2
```

```python
stacked = stack_df_2.stack()
```

```python
stacked
```

With a “stacked” DataFrame or Series (having a `MultiIndex` as the index), the inverse operation of `stack()` is `unstack()`D, which by default unstacks the last level

```python
stacked.unstack()
```

```python
stacked.unstack()
```

### Pivot Tables
`pivot_table()` pivots a `DataFrame` specifying the `values`, `index` and `columns`.

```python
pt_df = pd.DataFrame(
    {
        "A": ["one", "one", "two", "three"] * 3,
        "B": ["A", "B", "C"] * 4,
        "C": ["foo", "bar", "foo", "bar", "bar", "foo"] * 2,
        "D": np.random.randn(12),
        "E": np.random.randn(12),

    }
                    )
```

```python
pt_df
```

```python
pd.pivot_table(pt_df, values="D", index=["A", "B"], columns=["C"])
```
